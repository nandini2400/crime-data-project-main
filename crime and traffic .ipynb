
# 2. Data Cleaning and Transformation

#  Standardize Column Names

```python
crime_data.columns = crime_data.columns.str.lower().str.replace(' ', '_')
traffic_data.columns = traffic_data.columns.str.lower().str.replace(' ', '_')
```

# Extract Date and Time Components

```python
crime_data['date_occ'] = pd.to_datetime(crime_data['date_occ'], errors='coerce')
crime_data['year'] = crime_data['date_occ'].dt.year
crime_data['month'] = crime_data['date_occ'].dt.month
crime_data['day'] = crime_data['date_occ'].dt.day
crime_data['hour'] = crime_data['time_occ'].apply(lambda x: int(str(x).zfill(4)[:2]))
```

Add Derived columns 

```python
crime_data['is_weekend'] = crime_data['date_occ'].dt.dayofweek >= 5
```

# 3. Data Enrichment with External Data

Merge Datasets

```python
# Convert both date columns to date-only formats for approximate joins
traffic_data['crash_date'] = pd.to_datetime(traffic_data['crash_date'], errors='coerce')

# Merge based on date and location (e.g., within the same day and approximate location if latitude/longitude is available)
merged_data = pd.merge_asof(
    crime_data.sort_values('date_occ'),
    traffic_data.sort_values('crash_date'),
    left_on='date_occ',
    right_on='crash_date',
    direction='nearest'
)

# Check the merged data
display(merged_data.head(5))
```

# 4. ETL Pipeline Simulation

# 6. Visualization

Crime by Day of the Week and Hour (Heatmap)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pivot table for day and hour analysis
heatmap_data = merged_data.pivot_table(index='day', columns='hour', aggfunc='size', fill_value=0)
sns.heatmap(heatmap_data, cmap='YlGnBu')
plt.title('Crime by Day of the Week and Hour')
plt.show()
```
